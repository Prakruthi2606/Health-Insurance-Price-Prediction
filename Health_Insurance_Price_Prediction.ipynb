The descriptions of the variables based on the insurance dataset:

1. **Age**

2. **Sex**
   - **Description:** Gender of the insurance contractor.

   - **Type:** Discrete numerical variable.

   - **Description:** Binary indicator if the beneficiary is a smoker.
   - **Type:** Categorical variable with two levels: yes, no.

6. **Region**
   - **Description:** The beneficiary's residential area in the US.
   - **Type:** Categorical variable with four levels: northeast, southeast, southwest, northwest.

7. **Charges**
   - **Description:** Individual medical costs billed by health insurance.
   - **Type:** Continuous numerical variable.

### **Handling Outliers**

outliers are extreme values which are far away from actual data. Eg: Age of sudents:[20,22,23,24,25,40,98,100,160]. in this case 98,100,160 are outliers

**How to check outliers**
* box plot: Tells about the existence and direction of outliers
* scatter plot
* histogram

**How to find outliers**
* IQR(Interquartile range): Used when data distribution is skewed.
* 3-sigma rule: Used when distribution is normal
* outliers can be handled only if % of outliers is very less or less than 5%

### **IQR** (should be used only when the data distribution is skewed)
steps:
1. Find Q1
2. Find Q3
3. Find IQR = Q3-Q1
4. Find lower_limit = Q1 - 1.5 x IQR
5. Find upper_limit = Q3 + 1.5 x IQR
6. Any value which is greater than upper limit and lesser than lower limit is considered as **Outliers**
7. calculate the outliers percentage
8. if percentage is less than 5%, replace with median

**Check outliers in numerical columns**

```python
fig,ax=plt.subplots(1,3,figsize=(5,2))
sns.boxplot(data,x='age',ax=ax[0])
sns.boxplot(data,x='bmi',ax=ax[1])
sns.boxplot(data,x='charges',ax=ax[2])
plt.tight_layout()
```

**IQR methods for charges**
since distribution is skewed

```python
Q1= data.loc[:,'charges'].quantile(0.25)
Q3 = data.loc[:,'charges'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
#filter values greater than upper_limit
data[data.loc[:,'charges']>upper_limit]
```

```python
#filter values lesser than lower_limit
data[data.loc[:,'charges']<lower_limit]
```

```python
#filter values greater than upper_limit , less then lower_limit
data[(data.loc[:,'charges']>upper_limit) | (data.loc[:,'charges']<lower_limit)]
```

```python
#calculate percentage
len(data[(data.loc[:,'charges']>upper_limit) | (data.loc[:,'charges']<lower_limit)])/len(data) * 100
```

since % of outlier are more than 5% we dont handle them, because there can be high charges for the aged people

## **3-sigma rule**
we use 3-sigma rule for bmi column because the data distribution is normal
1. find mean
2. find standard deviation
3. find lower limit = mean - 3 x std
4. find upper limit = mean + 3 x std
5. filter values greater than upper limit or less than lower limit is considered as an outliers
6. calculate the outliers percentage
7. if percentage is less than 5%, replace with median

```python
mean = data.loc[:,'bmi'].mean()
print(mean)
```

```python
std = data.loc[:,'bmi'].std()
print(std)
```

```python
lower_limit = mean - 3 * std
upper_limit = mean + 3 * std
```

```python
#filter values greater than upper_limit
data[data.loc[:,'bmi']>upper_limit]
```

```python
#filter values lesser than lower_limit
data[data.loc[:,'bmi']<lower_limit]
```

```python
#filter values greater than upper_limit , less then lower_limit
data[(data.loc[:,'bmi']>upper_limit) | (data.loc[:,'bmi']<lower_limit)]
```

```python
#calculate percentage
outliers = data[(data['bmi'] > upper_limit) | (data['bmi'] < lower_limit)]
percentage = (len(outliers) / len(data)) * 100
print(percentage)
```

since % of outliers was less than 5% we replace mean as distribution is normal

```python
# replace with mean
data.loc[data.loc[:,'bmi']>upper_limit,'bmi']=data.loc[:,'bmi'].mean()
```

```python
# varifying
data[data.loc[:,'bmi']>upper_limit]
```

### **Encoding Technique**
the process of converting text to numbers so a machine can understand.
* label encoder
* one hot encoder
* manual encoder
* target encoder
* ordinal encoder

1. Label Encoders:
converts texts to numbers.
* it assigns uniques numbers from 0 to each value based on alphabetical order <br>
* eg: yes - 1 and no -0 <br> red - 2 , green - 1, blue - 0<br> UG - 2, PG - 0, PHD - 1<br> label encoder **should not** be used when we need to consider order or priority

```python
# Smoker has values like yes or no so we will use label encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['smoker'] = le.fit_transform(data['smoker'])
data.dtypes
```
